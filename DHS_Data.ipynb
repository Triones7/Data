```python
#Import libraries.
import json, urllib
 
#Query the API for a list of countries.
countries_url = r'http://api.dhsprogram.com/rest/dhs/countries'
 
#Obtain and Parse the list into a Python Object.
req = urllib.request.urlopen(countries_url)
resp = json.loads(req.read())
my_data = resp['Data']
 
#Display a list of Country Code and names.
for c in my_data:
    print(c['DHS_CountryCode'] + ' : ' + c['CountryName'])
```

```python
#Import libraries.
import json, urllib
 
#Query the API for the particular data.
data_url = r'http://api.dhsprogram.com/rest/dhs/data/AM,ED_EDAT_W_PRI,AM2010DHS'
 
#Obtain and Parse the list into a Python Object.
req = urllib.request.urlopen(data_url)
resp = json.loads(req.read())
my_data = resp['Data']
 
#Display a list of each characteristic and their value.
for c in my_data:
    print(c['CharacteristicLabel'] + ' : ' + str(c['Value']))
```
